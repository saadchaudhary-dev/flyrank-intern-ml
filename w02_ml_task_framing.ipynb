## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
The primary success metric is Spearman Rank Correlation. Since the objective is to rank pages by refresh opportunity rather than predict an exact numeric value,
Spearman correlation measures how closely the predicted ranking matches the desired ordering. A higher positive correlation indicates better prioritization.

Code Cell
print("Success Metric: Spearman Rank Correlation")
```
